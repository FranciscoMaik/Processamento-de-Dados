```
path_deep_no_aug = "/content/drive/My Drive/Arquivos 10 classes/Deep Features/"
path_hand = "/content/drive/My Drive/Arquivos 10 classes/Handcrafted/New Whit Selected/"
```

```
def read_pickle(name):
    with (open(name, 'rb')) as openfile:
        while True:
            try:
                one_instance = pickle.load(openfile)
            except EOFError:
                break
    one_instance = np.asanyarray(one_instance)
    return one_instance
```

```
x_deep_train = read_pickle(path_deep_no_aug + "x_train_deep_features_1.pickle")
x_deep_test = read_pickle(path_deep_no_aug + "x_test_deep_features_1.pickle")
x_hand_train = read_pickle(path_hand + "handcrafted_features_treino_whit_selected.pickle")
x_hand_test = read_pickle(path_hand + "handcrafted_features_teste_whit_selected.pickle")
```

```
print("Deep Test: ",x_deep_test.shape)
print("Deep Train: ",x_deep_train.shape)
print("Hand Test: ",x_hand_test.shape)
print("Hand Train: ",x_hand_train.shape)
```

```
features_concat_train = np.concatenate((x_deep_train,x_hand_train), axis = 1)
features_concat_test = np.concatenate((x_deep_test,x_hand_test), axis = 1)
```
