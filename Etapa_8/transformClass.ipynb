```
import pickle
import numpy as np
```

```
path = "/content/drive/My Drive/Arquivos 10 classes/Pickles Sem Augmentation/"
```

```
def read_pickle(name):
    with (open(name, 'rb')) as openfile:
        while True:
            try:
                one_instance = pickle.load(openfile)
            except EOFError:
                break
    one_instance = np.asanyarray(one_instance)
    return one_instance
```
