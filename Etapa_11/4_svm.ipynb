# S05 E02: Classificação com Support Vector Machine

## The one with libs

```python
import pickle
from sklearn.svm import SVC
from sklearn.model_selection import GridSearchCV
from sklearn.metrics import accuracy_score, cohen_kappa_score, balanced_accuracy_score, make_scorer
import numpy as np
import time
import pandas as pd
```

```python
path = "/content/drive/My Drive/Handcraft Features/Handcratf RFE/"
```

## The one who loads the train and test sets 

```python
def read_pickle(name):
    with (open(name, 'rb')) as openfile:
        while True:
            try:
                one_instance = pickle.load(openfile)
            except EOFError:
                break
    one_instance = np.asanyarray(one_instance)
    return one_instance
```

```python
ls '/content/drive/My Drive/Handcraft Features/Handcratf RFE'
```

```python
x_test = read_pickle(path+'handcrafted_no_aug_with_select_rfe_test_1.pickle')
x_train = read_pickle(path+'handcrafted_no_aug_with_select_rfe_train_1.pickle')
y_test = read_pickle(path+'Cópia de handcrafted_labels_teste_no_aumentation.pickle')
y_train = read_pickle(path+'Cópia de handcrafted_labels_treino_no_aumentation.pickle')
```

```python
print(x_train.shape)
print(x_test.shape)
print(y_train.shape)
print(y_test.shape)
```

# Melhores parâmetros

```python
initial_time = time.time() 
classifier = SVC()
param_grid = {'C': [0.1,0.5,1.0,1.5,1.7,2.0], 'kernel': ['poly','rbf','sigmoid']}
scoring = {'accuracy': make_scorer(accuracy_score), 
           'balanced_accuracy': make_scorer(balanced_accuracy_score)}
search = GridSearchCV(classifier,
                      param_grid,
                      scoring,
                      cv = 5,
                      refit = 'accuracy',
                      n_jobs = -1)
search.fit(x_train, y_train)
print("Melhores parâmetros: ",search.best_params_)
finally_time = time.time()
print("Time: ", (finally_time - initial_time)/60)
```

## The one who trains the model

```python
initial_time = time.time()
classifier = SVC(kernel = 'rbf', C= 2.0)
classifier.fit(x_train,y_train)
finally_time = time.time()
print("Tempo de duração: ", (finally_time - initial_time)/60)
```

## The one who tests the model

```python
pred = classifier.predict(x_test)
```

## The one who chooses the optimal values for the hyperparameters

```python
print('Acuracia: ',accuracy_score(y_test,pred))
print('Kappa: ', cohen_kappa_score(y_test,pred))
```
