# S05 E01: Classificação com Random Forest

## The one with libs

```python
import pickle
from sklearn.ensemble import RandomForestClassifier
from sklearn.calibration import CalibratedClassifierCV
from sklearn.model_selection import GridSearchCV
from sklearn.metrics import accuracy_score, confusion_matrix, cohen_kappa_score, roc_auc_score, balanced_accuracy_score, make_scorer
import numpy as np
import time
import pandas as pd
path = "/content/drive/My Drive/Handcraft Features/Handcratf RFE/"
```

## The one who loads the train and test sets 

```python
def read_pickle(name):
    with (open(name, 'rb')) as openfile:
        while True:
            try:
                one_instance = pickle.load(openfile)
            except EOFError:
                break
    one_instance = np.asanyarray(one_instance)
    return one_instance
```

```python
ls '/content/drive/My Drive/Handcraft Features/Handcratf RFE'
```

```python
x_test = read_pickle(path+'handcrafted_no_aug_with_select_rfe_test_1.pickle')
x_train = read_pickle(path+'handcrafted_no_aug_with_select_rfe_train_1.pickle')
y_test = read_pickle(path+'Cópia de handcrafted_labels_teste_no_aumentation.pickle')
y_train = read_pickle(path+'Cópia de handcrafted_labels_treino_no_aumentation.pickle')
```

```python
print("Treino Features: ", x_train.shape)
print("Teste Features: ", x_test.shape)
print("Treino Labels: ", y_train.shape)
print("Teste Labels: ", y_test.shape)
```

## Parametro de acuracia

```python
initial_time = time.time()
classifier = RandomForestClassifier()
param_grid = {'n_estimators': [90,100,110,120,130,140,150,200]}
scoring = {'accuracy': make_scorer(accuracy_score), 
           'balanced_accuracy': make_scorer(balanced_accuracy_score)}
search = GridSearchCV(classifier,
                      param_grid,
                      scoring,
                      cv = 5,
                      refit = 'accuracy',
                      n_jobs = -1)
search.fit(x_train, y_train)
print("Melhores parâmetros: ",search.best_params_)
finally_time = time.time()
print("Duration of execution: ", (finally_time - initial_time)/60)
```

## The one who trains the model

```python
classifier = RandomForestClassifier(n_estimators=200)
classifier.fit(x_train,y_train)
```

## The one who tests the model

```python
pred = classifier.predict(x_test)
```

## The one who shows the results

```python
print('Acuracia: ',accuracy_score(y_test,pred))
print('Kappa: ', cohen_kappa_score(y_test,pred))
```
